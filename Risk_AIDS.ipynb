```python
from tkinter import *
from math import sqrt

def solver(a,b,c,x,y,z):

    D = 1.34+0.25*a+0.17*b+0.11*c+0.34*x+0.09*y-0.0003*z
    
    if D>3:
        text = "Высокий риск развития СПИДа,\nкоэффициент риска превышает 3"
    if 2<=D<=3:
        text = "Средний риск развития СПИДа,\nкоэффициент риска находится в пределах от 2 до 3"
    if 2>=D:
        text = "Низкий риск развития СПИДа,\nкоэффициент риска менее 2"
    return text

def inserter(value):

    output.delete("0.0","end")
    output.insert("0.0",value)    

def clear(event):
    """ Clears entry form """
    caller = event.widget
    caller.delete("0", "end")

def handler():

    try:
        a_val = float(a.get())
        b_val = float(b.get())
        c_val = float(c.get())
        x_val = float(c.get())
        y_val = float(c.get())
        z_val = float(c.get())
        inserter(solver(a_val, b_val, c_val, x_val, y_val, z_val))
    except ValueError:
        inserter("Некорректно заполнены поля программы")

root = Tk()
root.title("Risk of developing AIDS")
root.minsize(325,230)
root.resizable(width=False, height=False)


frame = Frame(root)
frame.grid()


a = Entry(frame, bg="azure", width=3)
a.grid(row=3,column=2, padx=(10,0))
a.bind("<FocusIn>", clear)
a_lab = Label(frame, text="1. Диагноз стадии ВИЧ, баллы.").grid(row=3,column=1)

b = Entry(frame, bg="azure", width=3)
b.bind("<FocusIn>", clear)
b.grid(row=5,column=2,padx=(10,0))
b_lab = Label(frame, text="2. Стадия туберкулеза, баллы.").grid(row=5, column=1)

c = Entry(frame, bg="azure", width=3)
c.bind("<FocusIn>", clear)
c.grid(row=7, column=2, padx=(10,0))
c_lab = Label(frame, text="3. Температура тела, баллы.").grid(row=7, column=1)

x = Entry(frame, bg="azure", width=3 )
x.grid(row=9,column=2,padx=(10,0))
x.bind("<FocusIn>", clear)
x_lab = Label(frame, text="4. Хронические обструктивные \n болезни легких (баллы).").grid(row=9,column=1)

y = Entry(frame, bg="azure", width=3)
y.bind("<FocusIn>", clear)
y.grid(row=11,column=2, padx=(10,0))
y_lab = Label(frame, text="5. Размер печени по методике \n М.Г. Курлова, сантиметры.").grid(row=11, column=1)

z = Entry(frame, bg="azure", width=3)
z.bind("<FocusIn>", clear)
z.grid(row=13, column=2, padx=(10,0))
z_lab = Label(frame, text="6. СД4, тыс.кл/мл.").grid(row=13, column=1)

but = Button(frame,  text="Рассчитать риск развития СПИДа", bg="orange red", command=handler).grid(row=3, column=3, columnspan=5 )

output = Text(frame, bg="gray95", font="Times 12", width=50, height=3)
output.grid(row=16, column=1, columnspan=7)

Text = Label (frame, text="ПРОГНОЗИРОВАНИЕ РИСКА РАЗВИТИЯ СПИДа \n У ОСУЖДЕННЫХ В ВОЗРАСТЕ ДО 40 ЛЕТ", font="Times 12" ).grid(row=1, column=1, columnspan=7 )
Text = Label (frame,  bg="snow", text="ПРАВИЛА ЗАПОЛНЕНИЕ ПОЛЕЙ ПРОГРАММЫ: \n\n 1. Диагноз стадии ВИЧ определяется в баллах (1 стадия – 0 баллов, 2 стадия – 1 балл, \n 3 стадия – 2 балла, 4 стадия  А-Б – 3 балла, 4 стадия Б-В – 4 балла). \n\n 2. Стадия туберкулеза определяется в баллах (отсутствие туберкулеза -0 баллов; \n туберкулез в анамнезе, метатуберкулезные изменения в легких – 1 балл; \n туберкулез грудных лимфоузлов – 2 балла; туберкулез очаговый – 3 балла ; \n туберкулез инфильтративный – 4 балла; туберкулез диссеминированный  – 5 баллов;\n туберкулез с плевритом – 6 баллов; туберкулез генерализованный – 7 баллов; \n рецидив туберкулеза – 8 баллов). \n\n 3. Наличие либо отсутствие повышенной температуры (0 – температура от 36,6 до \n 37,0 градусов, 1 – невысокая температура от 37,0 до 38,0 градусов; \n 2 балла повышенная температура выше 38,0). \n\n 4. Наличие/отсутствие хронической обструктивной болезни легких определяется в баллах \n (наличие – 1 балл, отсутствие – 0 баллов). \n\n 5. Размер печени определяется по методике М.Г. Курлова, сантиметры; \n\n6. СД4 определяется в тыс.кл/мл.", font="Times 11").grid(row=18, column=1, columnspan=7 )
Text = Label (frame,   text=" ").grid(row=12)
Text = Label (frame,   text=" ").grid(row=10)
Text = Label (frame,   text=" ").grid(row=20)
Text = Label (frame,   text=" ").grid(row=2)


Text = Label (frame,   text="").grid(row=4)
Text = Label (frame,   text="Результат исследования:", font="Times 12").grid(row=15, column=2)
Text = Label (frame,   text="").grid(row=6)
Text = Label (frame,   text=" ").grid(row=8)
Text = Label (frame,   text=" ").grid(row=10)
Text = Label (frame,   text=" ").grid(row=12)
Text = Label (frame,   text=" ").grid(row=14)
Text = Label (frame,   text=" ").grid(row=17)

root.mainloop()
```
